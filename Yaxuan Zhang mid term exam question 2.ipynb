```python
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily
```

```python
start = datetime(2022, 5, 1)
end = datetime(2022, 5, 31)
#my birthday is in May
```

```python
location = Point(36.66528200, 116.99491700, 38.80)
#my birth city is Jinan,Shandong,China and latitude,36.66528200,Longitude:116.99491700, altitude:38.80,source from https://www.altitude-maps.com/city/43_316,Jinan,Shandong,China
```

```python
data = Daily(location, start, end)
data = data.fetch()
```

```python
data
```

```python
T_base = 18
cdds = [max(temp - T_base, 0) for temp in data.tavg]
```

```python
total_cdd = sum(cdds)
```

```python
print("Total CDD:", total_cdd)
```
